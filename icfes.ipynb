```python


# Definir number_of_bins y labels
number_of_bins = 10  # Ajusta según lo necesites
labels = ['Puntaje global', 'Puntaje sin etnia']  # Asegúrate de que coincida con el número de datasets

# Asegúrate de que las listas de datos son correctas
data_sets = [df_icfes_v['punt_global'], puntaje_sin_etnia]  # Asegúrate de que 'puntaje_global' existe

# Concatenar los datos para encontrar el mínimo y máximo global
all_data = np.concatenate(data_sets)

# Rango del histograma
hist_range = (np.min(all_data), np.max(all_data))

# Crear histogramas binned
binned_data_sets = [
    np.histogram(d, range=hist_range, bins=number_of_bins)[0]
    for d in data_sets
]

# Encontrar máximos
binned_maximums = np.max(binned_data_sets, axis=1)
x_locations = np.arange(0, len(data_sets) * np.max(binned_maximums), np.max(binned_maximums))

# Crear edges y centros
bin_edges = np.linspace(hist_range[0], hist_range[1], number_of_bins + 1)
heights = np.diff(bin_edges)
centers = bin_edges[:-1] + heights / 2

# Plotear los histogramas
fig, ax = plt.subplots()
for x_loc, binned_data, label in zip(x_locations, binned_data_sets, labels):
    lefts = x_loc - 0.5 * binned_data
    ax.barh(centers, binned_data, height=heights, left=lefts, label=label)

ax.set_xticks(x_locations)
ax.set_xticklabels(labels)

ax.set_ylabel("Valores de los Datos")
ax.set_xlabel("Conjuntos de Datos")
ax.legend()  # Añade leyenda si tienes múltiples datasets

plt.show()
```

Se piuede apreciar en el histograma horizontal como el puntaje del icfes efectivamente se concentra entre apróximadamente 180 hasta 295, donde cuando se quitan los puntajes de las etnias sube un poco más los puntajes concentrados mayores a 250.
